```python
if op == 1:
    print("O total é: ",num1 + num2)
elif op == 2:
    print("O total é: ",num1 - num2)
elif op == 3:
      print("O total é: ",num1 / num2)
elif op == 4:
      print("O total é: ",num1 * num2)
else:
      print("Operador invalido")
```
